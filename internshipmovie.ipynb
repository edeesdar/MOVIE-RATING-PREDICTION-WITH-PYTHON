```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(newdf['tags'])
```

```python
cosine_sim = cosine_similarity(X)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    title = title.lower().strip()
    
    try:
        idx = df[df['Name'].str.lower().str.strip() == title].index[0]
    except IndexError:
        print(f"Movie '{title}' not found in the DataFrame.")
        return None

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11] 
    movie_indices = [i[0] for i in sim_scores]
    recommendations = df['Name'].iloc[movie_indices].values

    recommendations_df = pd.DataFrame(recommendations, columns=['Recommended Movies'])
    formatted_recommendations = recommendations_df.to_string(index=False, header=False)
    formatted_recommendations = '\n'.join(line.lstrip() for line in formatted_recommendations.split('\n'))
    
    return formatted_recommendations
```
